```python
def convert_to_json_1(df):
    result = {}  # El diccionario final que contendrá la estructura JSON
    
    for col in df.columns:
        if '_' in col:
            # Columnas con barra baja indican anidación
            parts = col.split('_')
            current_dict = result
            for part in parts[:-1]:
                current_dict = current_dict.setdefault(part, {})
            current_dict[parts[-1]] = df[col].to_list()
        elif pd.notna(df[col]).any():
            # Columnas sin barra baja y con al menos un valor no nulo son propiedades de primer nivel
            result[col] = df[col].to_list()

    return json.dumps(result, indent=2)
```

```python
def convert_to_json_2(df):
    
    # Eliminar las columnas que contienen solo NaN
    df = df.dropna(axis=1, how='all')

    # Inicializar el diccionario para almacenar el JSON final
    json_data = {}

    # Iterar sobre las columnas del DataFrame
    for col in df.columns:
        # Verificar si la columna contiene "_"
        if '_' in col:
            # Verificar si todos los valores en la columna son NaN
            if df[col].isnull().all():
                # En caso de que todos los valores sean NaN, indicar anidación
                keys = col.split('_')
                current_dict = json_data
                for key in keys[:-1]:
                    current_dict = current_dict.setdefault(key, {})
            else:
                # En caso contrario, agregar la propiedad al nivel actual del JSON
                keys = col.split('_')
                json_data[keys[-1]] = df[col].to_list()
        else:
            # Si no contiene "_", agregar la propiedad al nivel actual del JSON
            json_data[col] = df[col].to_list()

    # Convertir el diccionario a formato JSON
    json_result = json.dumps(json_data, indent=2)

    return json_result
```

```python
def convert_to_json_3(df):
    # Drop columns that contain only NaN values
    df = df.dropna(axis=1, how='all')

    # Initialize the dictionary to store the final JSON
    json_data = {}

    # Iterate over the DataFrame rows
    for index, row in df.iterrows():
        market_dict = {}
        interval_dict = {}
        tiers_list = []

        # Get values from columns
        schema_version = row['schemaVersion']
        extract_start_datetime = row['extractStartDateTime']
        market_iso_code = row['Market_isoCode']
        market_store_id = row['Market_storeId']
        interval_start = row['Interval_start']
        interval_end = row['Interval_end']
        average_ticket_currency = row['AverageTicket_currency']

        # Check if the column contains "_"
        for col in df.columns:
            if '_' in col:
                keys = col.split('_')
                if not pd.isnull(row[col]):
                    # Check if all values in the column are NaN
                    if keys[-1] == "isoCode" or keys[-1] == "storeId":
                        market_dict[keys[-1]] = row[col]
                    elif keys[-1] == "start" or keys[-1] == "end":
                        interval_dict[keys[-1]] = row[col]
                    elif keys[-1] == "tier" or keys[-1] == "amount":
                        tiers_list.append({keys[-1]: row[col]})
            else:
                json_data[col] = row[col]

        # Build the JSON structure
        market_dict["Market"] = {"isoCode": market_iso_code, "storeId": int(market_store_id)}
        interval_dict["Interval"] = {"start": interval_start, "end": interval_end}
        average_ticket_dict = {"currency": average_ticket_currency, "TiersList": tiers_list}
        market_dict["AverageTicket"] = average_ticket_dict

        # Add to the final JSON
        if "MarketsList" not in json_data:
            json_data["MarketsList"] = []
        json_data["MarketsList"].append({"Market": market_dict, "Interval": interval_dict, "AverageTicket": average_ticket_dict})

    # Convert the dictionary to JSON format
    json_result = json.dumps(json_data, indent=2)

    return json_result
```

```python
def convert_to_json_4(df):
    # Drop columns that contain only NaN values
    df = df.dropna(axis=1, how='all')

    # Initialize the dictionary to store the final JSON
    json_data = {}

    # Create a dictionary to store data for each Market
    markets_data = {}

    # Iterate over the DataFrame rows
    for index, row in df.iterrows():
        # Get values from columns
        schema_version = row['schemaVersion']
        extract_start_datetime = row['extractStartDateTime']
        market_iso_code = row['Market_isoCode']
        market_store_id = row['Market_storeId']
        interval_start = row['Interval_start']
        interval_end = row['Interval_end']
        average_ticket_currency = row['AverageTicket_currency']

        # Check if the column contains "_"
        for col in df.columns:
            if '_' in col:
                keys = col.split('_')
                if not pd.isnull(row[col]):
                    # Check if all values in the column are NaN
                    if keys[-1] == "isoCode" or keys[-1] == "storeId":
                        markets_data[index, keys[-1]] = row[col]
                    elif keys[-1] == "start" or keys[-1] == "end":
                        markets_data[index, keys[-1]] = row[col]
                    elif keys[-1] == "tier" or keys[-1] == "amount":
                        markets_data[index, keys[-1]] = row[col]
            else:
                json_data[col] = row[col]

        # Build the JSON structure for each Market
        market_dict = {
            "isoCode": markets_data.get((index, "isoCode"), ""),
            "storeId": int(markets_data.get((index, "storeId"), 0))
        }
        interval_dict = {
            "start": markets_data.get((index, "start"), ""),
            "end": markets_data.get((index, "end"), "")
        }
        tier_dict = {
            "tier": markets_data.get((index, "tier"), 0),
            "amount": markets_data.get((index, "amount"), 0)
        }

        # Add to the final JSON
        if "MarketsList" not in json_data:
            json_data["MarketsList"] = []

        json_data["MarketsList"].append({
            "Market": market_dict,
            "Interval": interval_dict,
            "AverageTicket": {
                "currency": average_ticket_currency,
                "TiersList": [tier_dict]
            }
        })

    # Convert the dictionary to JSON format
    json_result = json.dumps(json_data, indent=2)
    
    return json_result
```

```python
def convert_to_json_5(df):
    # Drop columns that contain only NaN values
    df = df.dropna(axis=1, how='all')

    # Initialize the dictionary to store the final JSON
    json_data = {}

    # Create a dictionary to store data for each Market
    markets_data = {}

    # Iterate over the DataFrame rows
    for index, row in df.iterrows():
        # Get values from columns
        schema_version = row['schemaVersion']
        extract_start_datetime = row['extractStartDateTime']
        market_iso_code = row['Market_isoCode']
        market_store_id = row['Market_storeId']
        interval_start = row['Interval_start']
        interval_end = row['Interval_end']
        average_ticket_currency = row['AverageTicket_currency']

        # Check if the column contains "_"
        for col in df.columns:
            if '_' in col:
                keys = col.split('_')
                if not pd.isnull(row[col]):
                    # Check if all values in the column are NaN
                    if keys[-1] == "isoCode" or keys[-1] == "storeId":
                        markets_data[index, keys[-1]] = row[col]
                    elif keys[-1] == "start" or keys[-1] == "end":
                        markets_data[index, keys[-1]] = row[col]
                    elif keys[-1] == "tier" or keys[-1] == "amount":
                        markets_data[index, keys[-1]] = row[col]
            else:
                json_data[col] = row[col]

        # Build the JSON structure for each Market
        market_dict = {
            "isoCode": markets_data.get((index, "isoCode"), ""),
            "storeId": int(markets_data.get((index, "storeId"), 0))
        }
        interval_dict = {
            "start": markets_data.get((index, "start"), ""),
            "end": markets_data.get((index, "end"), "")
        }
        tier_dict = {
            "tier": markets_data.get((index, "tier"), 0),
            "amount": markets_data.get((index, "amount"), 0)
        }

        # Add to the final JSON
        if "MarketsList" not in json_data:
            json_data["MarketsList"] = []

        # Check if the current Market is already in the list
        market_exists = any(market["Market"] == market_dict for market in json_data["MarketsList"])

        if not market_exists:
            json_data["MarketsList"].append({
                "Market": market_dict,
                "Interval": interval_dict,
                "AverageTicket": {
                    "currency": average_ticket_currency,
                    "TiersList": [tier_dict]
                }
            })
        else:
            # If the Market already exists, append only the Tier to the existing Market
            existing_market = next(market for market in json_data["MarketsList"] if market["Market"] == market_dict)
            existing_market["AverageTicket"]["TiersList"].append(tier_dict)

    # Convert the dictionary to JSON format
    json_result = json.dumps(json_data, indent=2)

    return json_result
```
